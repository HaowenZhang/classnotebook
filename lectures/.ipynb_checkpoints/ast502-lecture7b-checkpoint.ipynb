<h2>Determining Ly-$\alpha$ Position for MAVEN/IUVS Calibration</h2>
<p>Although the spectral scale for IUVS is known approximately, various factors (e.g. temperature of instrument) can change the location of the image of Ly-$\alpha$ on the detector. The location of the Ly-$\alpha$ line is determined to refine calibration of each image.</p>

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
import csv

# Read in CSV data file
data = np.zeros((41,21))
with open('ast502-lecture7b-IUVSLya.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    i_int = 0
    for row in csvReader:
        for i_spe in range(len(row)):
            data[i_spe,i_int] = float(row[i_spe])
        i_int += 1

# Fit Gaussians to each profile
def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))
p0 = [3000., 22., 4.]

centroids = np.zeros((21,2))

for i_int in range(21):
    coeff, var_matrix = curve_fit(gauss, np.linspace(0,40,41), data[:,i_int], p0=p0)
    centroids[i_int,0] = coeff[1]
    centroids[i_int,1] = coeff[2]

Lp = np.zeros(1000)
i = 0
x = np.linspace(0,40,1000)
for mu in x:
    Lp[i] = -sum(np.divide(np.power((centroids[:,0]-mu),2.0)/2,np.power(centroids[:,1],2.0)))
    i += 1
    
Lp_max = np.amax(Lp)
x_max = x[Lp.argmax()]

fig = plt.figure(figsize=(12, 5))
fig.subplots_adjust(left=0.1, right=0.95, wspace=0.4,
                    bottom=0.1, top=0.95, hspace=0.2)

if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=24, usetex=True)

# Plot data
ax1 = fig.add_subplot(1,2,1)
for i_int in range(21):
    ax1.plot(range(41),data[:,i_int])

ax1.set_title('Data')
ax1.set_ylabel('Counts')
ax1.set_xlabel('Spectral Bin')

# Plot likelihoods
ax2 = fig.add_subplot(1,2,2)
ax2.plot(x,Lp)
ax2.vlines(x_max,-400,0,linestyles="dashed")

ax2.set_title('Likelihood')
ax2.set_ylabel('$L_p$')
ax2.set_xlabel('$\mu$')

plt.show()
```

<h2>Uniform Distribution</h2>
<p>Let us have a uniform distribution centered at an unknown $\mu$ with an unknown width $W$. Then the likelihood of observing $x_i$ is </p>
<p style="text-align:center">$p(x_i|\mu,W)=\begin{cases}1/W & \text{for }|x_i-\mu|\leq W/2\\0 & \text{otherwise}\end{cases}$</p>

<p>For $\mu$, we use a uniform prior between $-\Delta$ and $\Delta$, where $\Delta$ is much larger than the plausible range for $|\mu+W|$. </p>

<p>For $W$, we use a scale-invariant prior. </p>

<p>The posterior for $N$ observations is then</p>

<p style="text-align:center">$p(\mu,W|\{x_i\},I) \propto p(\{x_i\}|\mu,W,I)p(\mu|I)p(W|I)=\frac{1}{W^{N+1}}\prod^N_{i=1}g(\mu,W|x_i)$</p>

<p>where</p>

<p style="text-align:center">$g(\mu,W|x_i)=\begin{cases}1 & \text{for } |x_i-\mu|\leq W/2\\ 0 & \text{otherwise}\end{cases}$</p>


<p>Logarithm of the posterior probability distribution for $N=100$, $\mu=5$ and $W=10$. </p>
<p>Maximum probablity is normalized to 0. Bottom panel shows the marginal posterior for $\mu$ while left panel shows the marginal posterior for $W$. MAP is at $\mu=(x_{min}+x_{max})/2$, $W=x_{max}-x_{min}$.</p>

```python
# Obtained from http://www.astroml.org/book_figures/chapter5/fig_likelihood_uniform.html#book-fig-chapter5-fig-likelihood-uniform
from __future__ import print_function

import numpy as np
from matplotlib import pyplot as plt

if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=24, usetex=True)


def uniform_logL(x, W, mu):
    """Equation 5.76:"""
    xmin = np.min(x)
    xmax = np.max(x)
    n = x.size

    res = np.zeros(mu.shape, dtype=float) - (n + 1) * np.log(W)
    res[(abs(xmin - mu) > 0.5 * W) | (abs(xmax - mu) > 0.5 * W)] = -np.inf

    return res

#------------------------------------------------------------
# Define the grid and compute logL
W = np.linspace(9.7, 10.7, 70)
mu = np.linspace(4.5, 5.5, 70)

np.random.seed(0)
x = 10 * np.random.random(100)

logL = uniform_logL(x, W[:, None], mu)
logL -= logL.max()

#------------------------------------------------------------
# Compute marginal likelihoods
n = x.size

p_mu = np.exp(logL).sum(0)
Wmin = x.max() - x.min()
p_W = (W - Wmin) / W ** (n + 1)
p_W[W < Wmin] = 0
p_W /= p_W.sum()

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(15, 6))

# 2D likelihood plot
ax = fig.add_axes([0.35, 0.35, 0.45, 0.6], xticks=[], yticks=[])
logL[logL < -10] = -10  # truncate for clean plotting
plt.imshow(logL, origin='lower',
           extent=(mu[0], mu[-1], W[0], W[-1]),
           cmap=plt.cm.binary,
           aspect='auto')

# colorbar
cax = plt.axes([0.82, 0.35, 0.02, 0.6])
cb = plt.colorbar(cax=cax)
cb.set_label(r'$\log L(\mu, W)$')
plt.clim(-7, 0)

ax.set_xlim(4.5, 5.5)
ax.set_ylim(9.7, 10.7)

ax1 = fig.add_axes([0.35, 0.1, 0.45, 0.23], yticks=[])
ax1.plot(mu, p_mu, '-k')
ax1.set_xlabel(r'$\mu$')
ax1.set_ylabel(r'$p(\mu)$')
ax1.set_xlim(4.5, 5.5)

ax2 = fig.add_axes([0.15, 0.35, 0.18, 0.6], xticks=[])
ax2.plot(p_W, W, '-k')
ax2.set_xlabel(r'$p(W)$')
ax2.set_ylabel(r'$W$')
ax2.set_xlim(ax2.get_xlim()[::-1])  # reverse x axis
ax2.set_ylim(9.7, 10.7)

plt.show()
```

<p>The marginal posterior pdf for $\mu$ is </p>
<p style="text-align:center">$p(\mu)=\int_0^\infty p(\mu,W|\{x_i\},I)dW \propto (|\mu-\bar\mu|+W^-/2)^-N$</p>

<p>The marginal posterior pdf for $W$ is </p>
<p style="text-align:center">$p(\mu)=\int_{-\Delta}^\Delta p(\mu,W|\{x_i\},I)d\mu \propto \frac{W-W^-}{W^{N+1}}$ for $W\geq W^-$</p>

<h2>Nuisance Parameter: Gaussian + Uniform Background</h2>

<p>Let us have a mixture of a Gaussian distribution (quantity of interest) of known $\mu$ but unknown $A$ and $\sigma$, and a uniform distribution (nuisance parameter) of sufficiently large unknown $W$. The likelihood of obtaining a measurement $x_i$ is</p>

<p style="text-align:center">$p(x_i|A,\mu,\sigma)=\begin{cases}\frac{A}{\sqrt{2\pi}\sigma}\exp\left(\frac{-(x_i-\mu)^2}{2\sigma^2}\right)+\frac{1-A}{W} & \text{for } 0\leq x_i \leq W \\ 0 & \text{otherwise}\end{cases}$ </p>

<p>We adopt uniform prior distributions for $A$ and $\sigma$ and require that they are both non-negative. </p>


<p>Logarithm of the posterior probability density function for $N=200$, $\mu=5$, $\sigma=1$, $A=0.5$ and $W-10$. Maximum probablity is normalized to 0. Contours enclose regions containing 0.683, 0.955 and 0.997 of the cumulative posterior probability. </p>
<p>Covariance between $A$ and $\sigma$: increase of $\sigma$ will widen the Gaussian so that its tails would partially compensate for the strength of the background component.</p>

```python
# Obtained from http://www.astroml.org/book_figures/chapter5/fig_likelihood_gausslin.html#book-fig-chapter5-fig-likelihood-gausslin
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import truncnorm, uniform
from astroML.plotting.mcmc import convert_to_stdev

if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=24, usetex=True)


def gausslin_logL(xi, A=0.5, sigma=1.0, mu=5.0, L=10.0):
    """Equation 5.80: gaussian likelihood with uniform background"""
    xi = np.asarray(xi)

    shape = np.broadcast(sigma, A, mu, L).shape

    xi = xi.reshape(xi.shape + tuple([1 for s in shape]))

    return np.sum(np.log(A * np.exp(-0.5 * ((xi - mu) / sigma) ** 2)
                         / (sigma * np.sqrt(2 * np.pi))
                         + (1. - A) / L), 0)

#------------------------------------------------------------
# Define the distribution
np.random.seed(0)
mu = 5.0
sigma = 1.0
L = 10.0
A = 0.5
N = 200

xi = np.random.random(N)
NA = np.sum(xi < A)

dist1 = truncnorm((0 - mu) / sigma, (L - mu) / sigma, mu, sigma)
dist2 = uniform(0, 10)

xi[:NA] = dist1.rvs(NA)
xi[NA:] = dist2.rvs(N - NA)

x = np.linspace(-1, 11, 1000)
fracA = NA * 1. / N

#------------------------------------------------------------
# define the (sigma, A) grid and compute logL
sigma = np.linspace(0.5, 2, 70)
A = np.linspace(0, 1, 70)

logL = gausslin_logL(xi, A[:, np.newaxis], sigma)
logL -= logL.max()

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(7, 4.5))
#fig.subplots_adjust(bottom=0.07, left=0.11, hspace=0.15, top=0.95)

ax = fig.add_axes([0,0,1,1])
plt.imshow(logL, origin='lower', aspect='auto',
           extent=(sigma[0], sigma[-1], A[0], A[-1]),
           cmap=plt.cm.binary)
plt.colorbar().set_label(r'$\log(L)$')
plt.clim(-5, 0)
ax.set_xlabel(r'$\sigma$')
ax.set_ylabel(r'$A$')

ax.contour(sigma, A, convert_to_stdev(logL),
           levels=(0.683, 0.955, 0.997),
           colors='k')

plt.show()
```

<h2>Regression: Fitting Pairs of Variables</h2>

<p>Let us have a frequency distribution made up of pairs of ($x_i$, $y_i$), where $y$ correspond to counts in bins with centers at $x$ and a constant bin width $\Delta$. Thus we have a dataset $\{x_i\}$ of size $N=\sum_{j=1}^M y_j$. We want to estimate $a$ and $b$ for the model $y=ax+b$.</p>

<h3>No Binning</h3>
<p>First we assume that we know the exact $x$ values for individual measurement, ie $\Delta = 0$. Then we have </p>
<p style="text-align:center">$p(x)= \begin{cases} ax+b & \text{for } x_{min} \leq x \leq x_{max}\\ 0 & \text{otherwise}\end{cases}$</p>
<p>We also define </p>

<p style="text-align:center">$W=x_{max}-x_{min}$</p>
<p style="text-align:center">$x_{1/2}=\frac{1}{2}(x_{min}+x_{max})$. </p>

<p>For $p(x)$ to be normalized, </p>

<p style="text-align:center">$b=1/W-ax_{1/2}$</p>

<p>The normalized pdf is then</p>

<p style="text-align:center">$p(x_i|a)=a(x_i-x_{1/2})+1/W$</p>

<p>Likelihood for different values of $a$ for three different sample sizes. The input value (0.01) is indicated by the vertical dotted line.</p>
<p>The distribution is sharper and more centered at 0.01 with more data points.</p>

```python
# Obtained from http://www.astroml.org/book_figures/chapter5/fig_poisson_continuous.html#book-fig-chapter5-fig-poisson-continuous

import numpy as np
from matplotlib import pyplot as plt
from astroML.stats.random import linear

if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=24, usetex=True)


def linprob_logL(x, a, xmin, xmax):
    x = x.ravel()
    a = a.reshape(a.shape + (1,))

    mu = 0.5 * (xmin + xmax)
    W = (xmax - xmin)

    return np.sum(np.log(a * (x - mu) + 1. / W), -1)

#----------------------------------------------------------------------
# Draw the data from the linear distribution
np.random.seed(0)

N = 500

a_true = 0.01
xmin = 0.0
xmax = 10.0

lin_dist = linear(xmin, xmax, a_true)
data = lin_dist.rvs(N)
x = np.linspace(xmin - 1, xmax + 1, 1000)
px = lin_dist.pdf(x)

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(7, 4.5))

ax = fig.add_axes([0,0,1,1])
ax.xaxis.set_major_locator(plt.MultipleLocator(0.01))

a = np.linspace(-0.01, 0.02, 1000)
Npts = (500, 100, 20)
styles = ('-k', '--b', '-.g')

for n, s in zip(Npts, styles):
    logL = linprob_logL(data[:n], a, xmin, xmax)
    logL = np.exp(logL - logL.max())
    logL /= logL.sum() * (a[1] - a[0])

    ax.plot(a, logL, s, label=r'$\rm %i\ pts$' % n)
ax.legend(loc=2, prop=dict(size=24))

ax.set_xlim(-0.011, 0.02)

ax.set_xlabel('$a$')
ax.set_ylabel('$p(a)$')

# vertical line: in newer matplotlib versions, use ax.vlines([a_true])
ylim = ax.get_ylim()
ax.plot([a_true, a_true], ylim, ':k', lw=1)
ax.set_ylim(ylim)

plt.show()
```

<h3>With Binning</h3>
<p>With bin width $\Delta\neq 0$, the model for predicting the expected counts in a given bin is </p>
<p style="text-align:center">$y(x)=a^*x+b^*$</p>
<p>Note that $a^*$ and $b^*$ are different from $a$ and $b$ &#8212; they describe the number of measurements falling into a bin $x$ rather than the value of $x$. The actual counts $y_i$ are "scattered around" their true values $\mu_i=a^*x_i+b^*$.</p>

<p>If $\mu_i$ are sufficiently large, the data likelihood can be approximated as a Gaussian, giving us</p>
<p style="text-align:center">$L_p^G(a^*x,b^*) \equiv \ln[p(a^*x,b^*|\{x_i,y_i\})]$ </p>
<p style="text-align:center">$= \text{constant}-\frac{1}{2}\sum_{i=1}^M\left[\ln(a^*x_i+b^*)+\frac{(y_i-a^*x_i-b^*)^2}{a^*x_i+b}\right]$</p>

<p> If $\mu_i$ is not large, then a Poisson distribution is more appropriate: </p>
<p style="text-align:center">$L_p^P(a^*x,b^*) = \text{constant}+\sum_{i=1}^M[y_i\ln(a^*x_i+b^*)-a^*x_i-b^*]$</p>

<p>1$\sigma$, 2$\sigma$ and 3$\sigma$ likelihood contours for the Poisson and Gaussian models.</p>
<p>When counts are few, the Gaussian likelihoods deviate more from the Poisson likelihoods. The expectation value and standard deviation for $y_i$ remain the same across both cases. The major axis of the likelihood contours corresponds to the normalization constraint in the $\Delta=0$ case. The binning process resulted in a "blurring" of the constraint.</p>

```python
# Obtained from http://www.astroml.org/book_figures/chapter5/fig_poisson_likelihood.html
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats, interpolate
from astroML.stats.random import linear
from astroML.plotting.mcmc import convert_to_stdev

if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=24, usetex=True)


def logL_gaussian(xi, yi, a, b):
    """gaussian log-likelihood (Eq. 5.87)"""
    xi = xi.ravel()
    yi = yi.ravel()
    a = a.reshape(a.shape + (1,))
    b = b.reshape(b.shape + (1,))
    yyi = a * xi + b
    return -0.5 * np.sum(np.log(yyi) + (yi - yyi) ** 2 / yyi, -1)


def logL_poisson(xi, yi, a, b):
    """poisson log-likelihood (Eq. 5.88)"""
    xi = xi.ravel()
    yi = yi.ravel()
    a = a.reshape(a.shape + (1,))
    b = b.reshape(b.shape + (1,))
    yyi = a * xi + b

    return np.sum(yi * np.log(yyi) - yyi, -1)


#------------------------------------------------------------
# Draw points from distribution
np.random.seed(0)

N = 50

a_true = 0.01
xmin = 0.0
xmax = 10.0
b_true = 1. / (xmax - xmin) - 0.5 * a_true * (xmax + xmin)

lin_dist = linear(xmin, xmax, a_true)
data = lin_dist.rvs(N)

#------------------------------------------------------------
# Compute and plot the results
fig = plt.figure(figsize=(12, 5))
fig.subplots_adjust(left=0.1, right=0.95, wspace=0.4,
                    bottom=0.1, top=0.95, hspace=0.2)

a = np.linspace(0.00001, 0.04, 71)
b = np.linspace(0.00001, 0.15, 71)

for num, nbins in enumerate([5, 40]):
    
    # divide points into bins
    yi, bins = np.histogram(data, bins=np.linspace(xmin, xmax, nbins + 1))
    xi = 0.5 * (bins[:-1] + bins[1:])

    # compute likelihoods for Poisson and Gaussian models
    factor = N * (xmax - xmin) * 1. / nbins
    LP = logL_poisson(xi, yi, factor * a, factor * b[:, None])
    LG = logL_gaussian(xi, yi, factor * a, factor * b[:, None])

    LP -= np.max(LP)
    LG -= np.max(LG)

    # find maximum likelihood point
    i, j = np.where(LP == np.max(LP))
    aP, bP = a[j[0]], b[i[0]]

    i, j = np.where(LG == np.max(LG))
    aG, bG = a[j[0]], b[i[0]]

    # plot likelihood contours
    ax = fig.add_subplot(1, 2, 1 + num)

    ax.contour(a, b, convert_to_stdev(LP),
               levels=(0.683, 0.955, 0.997),
               colors='k', linewidths=2)

    ax.contour(a, b, convert_to_stdev(LG),
               levels=(0.683, 0.955, 0.997),
               colors='b', linewidths=2)
    
    ax.set_title(r'$\rm %i\ points$' % N + r', $\rm %i\ bins$' % nbins)

    # plot horizontal and vertical lines
    #  in newer matplotlib versions, use ax.vlines() and ax.hlines()
    ax.plot([a_true, a_true], [0, 0.2], ':k', lw=1)
    ax.plot([0, 0.06], [b_true, b_true], ':k', lw=1)

    ax.set_xlabel(r'$a^\ast$')
    ax.set_ylabel(r'$b^\ast$')

    ax.set_xlim(0, 0.04)
    ax.set_ylim(0.001, 0.15)

    ax.xaxis.set_major_locator(plt.MultipleLocator(0.02))

# trick the legend command
ax.plot([0], [0], '-k', lw=2, label='Poisson')
ax.plot([0], [0], '-', c='b', lw=2, label='Gaussian')
ax.legend(loc=0)
    
plt.show()
```

<h2>Modeling Bad Data Points</h2>

<p>Assume we have a set of $N$ measurements, $\{x_i\}$, of a single-valued quantity $\mu$, with Gaussian, heteroscedastic and known measurement errors $\sigma_i$. We can model all the bad data points with $p_{bad}(x_i|\mu,I)$, with the "switching" variable $g_i$. The data likelihood is then</p>

<p style="text-align:center">$p(x_i|\mu,\sigma_i,g_i,I)=g_i\mathcal{N}(\mu,\sigma_i)+(1-g_i)p_{bad}(x_i|\mu,I)$</p>
<p>where $g_i=1$ if the data point is good and $g_i=0$ if it is bad. While we do not know the values of $g_i$, we can treat them as model parameters and then marginalize over them to get $p(\mu,\{x_i\},\{\sigma_i\},I)$. Assuming a separable prior</p>

<p style="text-align:center">$p(\mu,\{g_i\}|I)=p(\mu|I)p(\{g_i\}|I)$</p>

<p>The posterior is then</p>
<p style="text-align:center">$p(\mu,\{g_i\}|\{x_i\},I)\propto p(\mu,\{g_i\},I)p(\{x_i\}|\mu,\{\sigma_i\},\{g_i\},I)$</p>

<p style="text-align:center">$=p(\mu|I)p(\{g_i\}|I)\prod_{i=1}^N[g_i\mathcal{N}(\mu,\sigma_i)+(1-g_i)p_{bad}(x_i|\mu,I)]$</p>

<p>For simplicity we can use uniform priors for all $g_i$ ($g_i=0.5$) and a wide Gaussian for the bad data points</p>
<p style="text-align:center">$p_{bad}(x_i|\mu,I)=\mathcal{N}(\mu,\sigma^*)$</p>
<p>Then marginalizing over all $g_i$ we get</p>
<p style="text-align:center">$p(\mu|\{x_i\},\{\sigma_i\})\propto\prod_{i=1}^N[\mathcal{N}(\mu,\sigma_i)+\mathcal{N}(\mu,\sigma^*)]$</p>
<p>To determine if some data point $g_1$ is good, we can marginalize over all $g_{i\neq 1}$:</p>
<p style="text-align:center">$p(\mu,g_1|\{x_i\},\{\sigma_i\})$</p>
<p style="text-align:center">$\propto[g_1\mathcal{N}(\mu,\sigma_1)+(1-g_1)\mathcal{N}(\mu,\sigma^*)]\prod_{i=2}^N[\mathcal{N}(\mu,\sigma_i)+\mathcal{N}(\mu,\sigma^*)]$</p>

<p>The marginal joint distribution between $\mu$ and $g_i$. </p>

<p>We see that the bad point has a maximum at $g_1=0$, and as $g_1$ increases, it skews $\mu$ towards higher values. The good point has a maximum at $g_1=1$. 

```python
# Obtained from http://www.astroml.org/book_figures/chapter5/fig_outlier_likelihood.html
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
from astroML.plotting.mcmc import convert_to_stdev

if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=24, usetex=True)


def p(mu, g1, xi, sigma1, sigma2):
    """Equation 5.97: marginalized likelihood over outliers"""
    L = (g1 * norm.pdf(xi[0], mu, sigma1) +
         (1 - g1) * norm.pdf(xi[0], mu, sigma2))

    mu = mu.reshape(mu.shape + (1,))
    g1 = g1.reshape(g1.shape + (1,))

    return L * np.prod(norm.pdf(xi[1:], mu, sigma1)
                       + norm.pdf(xi[1:], mu, sigma2), -1)

#------------------------------------------------------------
# Sample the points
np.random.seed(138)

N1 = 8
N2 = 2
sigma1 = 1
sigma2 = 3

sigmai = np.zeros(N1 + N2)
sigmai[N2:] = sigma1
sigmai[:N2] = sigma2

xi = np.random.normal(0, sigmai)

#------------------------------------------------------------
# Compute the marginalized posterior for the first and last point
mu = np.linspace(-5, 5, 71)
g1 = np.linspace(0, 1, 11)

L1 = p(mu[:, None], g1, xi, 1, 10)
L1 /= np.max(L1)

L2 = p(mu[:, None], g1, xi[::-1], 1, 10)
L2 /= np.max(L2)

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(10, 4.5))
fig.subplots_adjust(left=0.1, right=0.95, wspace=0.15,
                    bottom=0.15, top=0.9)

ax1 = fig.add_subplot(121)
ax1.imshow(L1.T, origin='lower', aspect='auto', cmap=plt.cm.binary,
           extent=[mu[0], mu[-1], g1[0], g1[-1]])
ax1.contour(mu, g1, convert_to_stdev(np.log(L1).T),
            levels=(0.683, 0.955, 0.997),
            colors='k')
ax1.set_xlabel(r'$\mu$')
ax1.set_ylabel(r'$g_1$')
ax1.set_title("Bad Data Point")

ax2 = fig.add_subplot(122)
ax2.imshow(L2.T, origin='lower', aspect='auto', cmap=plt.cm.binary,
           extent=[mu[0], mu[-1], g1[0], g1[-1]])
ax2.contour(mu, g1, convert_to_stdev(np.log(L2).T),
            levels=(0.683, 0.955, 0.997),
            colors='k')
ax2.set_xlabel(r'$\mu$')
ax2.yaxis.set_major_locator(plt.NullLocator())
ax2.set_title("Good Data Point")

plt.show()
```

<p>We can also marginalize over $\mu$:</p>

```python
# Obtained from http://www.astroml.org/book_figures/chapter5/fig_outlier_marginalized.html
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=24, usetex=True)


def p(mu, g1, xi, sigma1, sigma2):
    """Equation 5.97: marginalized likelihood over outliers"""
    L = (g1 * norm.pdf(xi[0], mu, sigma1) +
         (1 - g1) * norm.pdf(xi[0], mu, sigma2))

    mu = mu.reshape(mu.shape + (1,))
    g1 = g1.reshape(g1.shape + (1,))

    return L * np.prod(norm.pdf(xi[1:], mu, sigma1)
                       + norm.pdf(xi[1:], mu, sigma2), -1)

#------------------------------------------------------------
# Sample the points
np.random.seed(138)

N1 = 8
N2 = 2
sigma1 = 1
sigma2 = 3

sigmai = np.zeros(N1 + N2)
sigmai[N2:] = sigma1
sigmai[:N2] = sigma2

xi = np.random.normal(0, sigmai)

#------------------------------------------------------------
# Compute the marginalized posterior for the first and last point
mu = np.linspace(-5, 5, 71)
g1 = np.linspace(0, 1, 71)

L1 = p(mu[:, None], g1, xi, 1, 10)
L1 /= np.max(L1)

(i1, j1) = np.where(L1 == 1)
mu0_1 = mu[i1]


L2 = p(mu[:, None], g1, xi[::-1], 1, 10)
L2 /= np.max(L2)
(i2, j2) = np.where(L2 == 1)
mu0_2 = mu[i2]

p1 = L1.sum(0)
p2 = L2.sum(0)

p1 /= np.sum(p1) * (g1[1] - g1[0])
p2 /= np.sum(p2) * (g1[1] - g1[0])

p1a = L1[i1[0]]
p2a = L2[i2[0]]

p1a /= p1a.sum() * (g1[1] - g1[0])
p2a /= p2a.sum() * (g1[1] - g1[0])

#------------------------------------------------------------
# Plot the results
fig, ax = plt.subplots(figsize=(10, 4.5))

l1, = ax.plot(g1, p1, '-k', lw=2)
l2, = ax.plot(g1, p1a, '--k', lw=2)
leg1 = ax.legend([l1, l2],
                 [r'$p(g_1)$ (bad point)',
                  r'$p(g_1|\mu_0)$ (bad point)'], loc=9, fontsize=18)

l3, = ax.plot(g1, p2, '-b', lw=1, label=r'$p(g_1)$ (good point)')
l4, = ax.plot(g1, p2a, '--b', lw=1, label=r'$p(g_1|\mu_0)$ (good point)')
leg2 = ax.legend([l3, l4],
                 [r'$p(g_1)$ (good point)',
                  r'$p(g_1|\mu_0)$ (good point)'], loc=8, fontsize=18)

# trick to display two legends:
# when legend() is called the second time, the first one is removed from
# the axes.  We add it back in here:
ax.add_artist(leg1)

ax.set_xlabel('$g_1$')
ax.set_ylabel('$p(g_1)$')

ax.set_xlim(0, 1)
ax.set_ylim(0.1, 1.8)

plt.show()
```

<p>For the good point, $g_1$ is more likely to be equal to 1. </p>
<p>For the bad point, $g_1$ is more likely to be equal to 0. When we fix the value of $\mu$ to be equal to the MAP estimate $\mu_0$, the bad point cannot skew the value of $\mu$ and appears even worse.</p>
