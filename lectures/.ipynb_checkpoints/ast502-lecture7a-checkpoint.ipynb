# Bayes' Theorem
\begin{aligned}
p(\theta | D) &=\frac{p(D | \theta) * p(\theta)}{p(D)} \\
\text {posterior} &=\frac{\text {likelihood} * \text { prior }}{\text {evidence}}
\end{aligned}

# Parameter Estimation for a Gaussian Distribution
## 1. Determining $\mu$ when measurement errors are known
Given N measurements $\{x_i\}$. The measurement errors, $\sigma_i$, are Gaussian and known for each measurement.

The likelihood function is the product of the likelihood of individual data points.


$$
p\left(\left\{x_{i}\right\} | \mu, I\right)=\prod_{i=1}^{N} \frac{1}{\sqrt{2 \pi} \sigma_{i}} \exp \left(\frac{-\left(x_{i}-\mu\right)^{2}}{2 \sigma_{i}^{2}}\right)
$$

Prior for $\mu$


$$
p(\mu | I)=C, \text { for } \mu_{\min }<\mu<\mu_{\max } \text {, where } C=\left(\mu_{\max }-\mu_{\min }\right)^{-1}
$$



The log of the posterior pdf of $\mu$ is 

$$\begin{align}
L_{p} &=\ln \left[p\left(\mu |\left\{x_{i}\right\},\left\{\sigma_{i}\right\}, I\right)\right] = \ln \left[p\left(\left\{x_{i}\right\} | \mu, I\right)\times p(\mu | I)\right]\\
&=\text { constant }-\sum_{i=1}^{N} \frac{\left(x_{i}-\mu\right)^{2}}{2 \sigma_{i}^{2}}
\end{align}
$$

Setting $\left.\left(d L_{p} / d \mu\right)\right|_{\left(\mu=\mu_{0}\right)}=0$, we get $\mu_0, \sigma_0$

$$
\mu_{0}=\frac{\sum_{i=1}^{N} w_{i} x_{i}}{\sum_{i=1}^{N} w_{i}}, 
\sigma_{\mu}=\left(-\left.\frac{d^{2} L_{p}}{d \mu^{2}}\right|_{\mu=\mu_{0}}\right)^{-1 / 2}=\left(\sum_{i=1}^{N} \frac{1}{\sigma_{i}^{2}}\right)^{-1 / 2}=\left(\sum_{i} w_{i}\right)^{-1 / 2}
$$

Posterior pdf of $\mu$ is Gaussian when $\sigma_i$ are known, regardless of the data size N.

## 2. Determining $\mu$ and $\sigma$
Data points are drawn from an unspecified Gaussian distribution $\mathcal{N}(\mu, \sigma)$.


Now, we instead need to find a two dimensional posterior distribution $p\left(\mu, \sigma |\left\{x_{i}\right\}\right)$

Assuming that the likelihood function for a single measurement $x_i$ follows a Gaussian distribution, the likelihood for all measurements is

$$
p\left(\left\{x_{i}\right\} | \mu, \sigma, I\right)=\prod_{i=1}^{N} \frac{1}{\sqrt{2 \pi} \sigma} \exp \left(\frac{-\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}\right)
$$

Same as earlier equation, only difference being that it is now a function of two parameters.

Choose a prior again


$$
p(\mu, \sigma | I) \propto \frac{1}{\sigma}, \text { for } \mu_{\min } \leq \mu \leq \mu_{\max } \text { and } \sigma_{\min } \leq \sigma \leq \sigma_{\max }
$$

The posterior distribution then turns out to be 

$$
p\left(\mu, \sigma |\left\{x_{i}\right\}, I\right) =p\left(\left\{x_{i}\right\} | \mu, \sigma, I\right) \times p(\mu, \sigma | I)=C \frac{1}{\sigma^{(N+1)}} \prod_{i=1}^{N} \exp \left(\frac{-\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}\right)
$$


where
$
C=(2 \pi)^{-N / 2}\left(\mu_{\max }-\mu_{\min }\right)^{-1}\left[\ln \left(\frac{\sigma_{\max }}{\sigma_{\min }}\right)\right]^{-1}
$

The logarithm of the posterior pdf is 

$$L_{p} \equiv \ln \left[p\left(\mu, \sigma |\left\{x_{i}\right\}, I\right)\right]=\text { constant }-(N+1) \ln \sigma-\sum_{i=1}^{N} \frac{\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}$$

$$L_{p} \equiv \ln \left[p\left(\mu, \sigma |\left\{x_{i}\right\}, I\right)\right]=\text { constant }-(N+1) \ln \sigma-\sum_{i=1}^{N} \frac{\left(x_{i}-\mu\right)^{2}}{2 \sigma^{2}}$$

Using 
$
\sum_{i=1}^{N}\left(x_{i}-\mu\right)^{2}=N(\bar{x}-\mu)^{2}+\sum_{i=1}^{N}\left(x_{i}-\bar{x}\right)^{2}
$, 

And with the substitution $V=N^{-1} \sum_{i=1}^{N}\left(x_{i}-\bar{x}\right)^{2}$

We can rewrite the posterior pdf in terms of data based quantities

$$
L_{p}=\text { constant }-(N+1) \ln \sigma-\frac{N}{2 \sigma^{2}}\left((\bar{x}-\mu)^{2}+V\right)
$$

Because of our assumption of a Gaussian distribution, the entire information content is captured by just three quantities $\bar{x}, V$ and $N$.

```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
import numpy as np
from matplotlib import pyplot as plt
from astroML.plotting.mcmc import convert_to_stdev

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=False)


def gauss_logL(xbar, V, n, sigma, mu):
    """Equation 5.57: gaussian likelihood"""
    return (-(n + 1) * np.log(sigma)
            - 0.5 * n * ((xbar - mu) ** 2 + V) / sigma ** 2)

#------------------------------------------------------------
# Define the grid and compute logL
sigma = np.linspace(1, 5, 70)
mu = np.linspace(-3, 5, 70)
xbar = 1
V = 4
n = 10

logL = gauss_logL(xbar, V, n, sigma[:, np.newaxis], mu)
logL -= logL.max()

#------------------------------------------------------------
def plot():
    # Plot the results
    fig = plt.figure(figsize=(7, 5))
    plt.imshow(logL, origin='lower',
               extent=(mu[0], mu[-1], sigma[0], sigma[-1]),
               cmap=plt.cm.binary,
               aspect='auto')
    plt.colorbar().set_label(r'$\log(L)$')
    plt.clim(-5, 0)

    plt.contour(mu, sigma, convert_to_stdev(logL),
                levels=(0.683, 0.955, 0.997),
                colors='k')

    plt.text(0.5, 0.93, r'$L(\mu,\sigma)\ \mathrm{for}\ \bar{x}=1,\ V=4,\ n=10$',
             bbox=dict(ec='k', fc='w', alpha=0.9),
             ha='center', va='center', transform=plt.gca().transAxes)

    plt.xlabel(r'$\mu$')
    plt.ylabel(r'$\sigma$')

    plt.show()
```

```python
plot()
```

## Calculating $p(\mu|x)$ and $p(\sigma|x)$

Marginilise over the other parameter

$$p\left(\mu |\left\{x_{i}\right\}, I\right)=\int_{0}^{\infty} p\left(\mu, \sigma |\left\{x_{i}\right\}, I\right) d \sigma \implies p\left(\mu |\left\{x_{i}\right\}, I\right) \propto\left[1+\frac{(\bar{x}-\mu)^{2}}{V}\right]^{-N / 2}$$

$$p\left(\sigma |\left\{x_{i}\right\}, I\right)=\int_{-\infty}^{\infty} p\left(\mu, \sigma |\left\{x_{i}\right\}, I\right) d \mu \implies p\left(\sigma |\left\{x_{i}\right\}, I\right) \propto \frac{1}{\sigma^{N}} \exp \left(\frac{-N V}{2 \sigma^{2}}\right)$$

```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
import numpy as np
from matplotlib import pyplot as plt
from astroML.stats import mean_sigma
from astroML.resample import bootstrap

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
#------------------------------------------------------------
# Define functions for computations below
#  These are expected analytic fits to the posterior distributions
def compute_pmu(mu, xbar, V, n):
    return (1 + (xbar - mu) ** 2 / V) ** (-0.5 * n)


def compute_pmu_alt(mu, xbar, V, n):
    return (1 + (xbar - mu) ** 2 / V) ** (-0.5 * (n - 1))


def compute_psig(sig, V, n):
    return (sig ** -n) * np.exp(-0.5 * n * V / sig ** 2)


def compute_psig_alt(sig, V, n):
    return (sig ** -(n - 1)) * np.exp(-0.5 * n * V / sig ** 2)


def gaussian(x, mu, sigma):
    return np.exp(-0.5 * (x - mu) ** 2 / sigma ** 2)


#------------------------------------------------------------
# Draw a random sample from the distribution, and compute
# some quantities
n = 10
xbar = 1
V = 4
sigma_x = np.sqrt(V)

np.random.seed(10)
xi = np.random.normal(xbar, sigma_x, size=n)

mu_mean, sig_mean = mean_sigma(xi, ddof=1)

# compute the analytically expected spread in measurements
mu_std = sig_mean / np.sqrt(n)
sig_std = sig_mean / np.sqrt(2 * (n - 1))

#------------------------------------------------------------
# bootstrap estimates
mu_bootstrap, sig_bootstrap = bootstrap(xi, int(1E6), mean_sigma,
                                        kwargs=dict(ddof=1, axis=1))

#------------------------------------------------------------
# Compute analytic posteriors

# distributions for the mean
mu = np.linspace(-3, 5, 1000)
dmu = mu[1] - mu[0]

pmu = compute_pmu(mu, 1, 4, n)
pmu /= (dmu * pmu.sum())

pmu2 = compute_pmu_alt(mu, 1, 4, n)
pmu2 /= (dmu * pmu2.sum())

pmu_norm = gaussian(mu, mu_mean, mu_std)
pmu_norm /= (dmu * pmu_norm.sum())

mu_hist, mu_bins = np.histogram(mu_bootstrap, 20)
mu_dbin = np.diff(mu_bins).astype(float)
mu_hist = mu_hist / mu_dbin / mu_hist.sum()

# distributions for the standard deviation
sig = np.linspace(1E-4, 8, 1000)
dsig = sig[1] - sig[0]
psig = compute_psig(sig, 4, n)
psig /= (dsig * psig.sum())

psig2 = compute_psig_alt(sig, 4, n)
psig2 /= (dsig * psig2.sum())

psig_norm = gaussian(sig, sig_mean, sig_std)
psig_norm /= (dsig * psig_norm.sum())

sig_hist, sig_bins = np.histogram(sig_bootstrap, 20)
sig_dbin = np.diff(sig_bins).astype(float)
sig_hist = sig_hist / sig_dbin / sig_hist.sum()

#------------------------------------------------------------
# Plot the results
def plot():
    fig = plt.figure(figsize=(10, 10))
    fig.subplots_adjust(wspace=0.35, right=0.95,
                        hspace=0.2, top=0.95)

    # plot posteriors for mu
    ax1 = plt.subplot(221, yscale='log')
    ax1.plot(mu, pmu, '-b')
    ax1.plot(mu, pmu2, ':m', label=r'unif. $\sigma$ prior')
    ax1.plot(mu, pmu_norm, '--r', label='Gaussian')

    ax1.set_xlabel(r'$\mu$')
    ax1.set_ylabel(r'$p(\mu|x,I)$')
    ax1.legend()

    ax2 = plt.subplot(223, sharex=ax1)
    ax2.plot(mu, pmu.cumsum() * dmu, '-b')
    ax2.plot(mu, pmu_norm.cumsum() * dmu, '--r')

    ax2.set_xlim(-3, 5)

    ax2.set_xlabel(r'$\mu$')
    ax2.set_ylabel(r'$P(<\mu|x,I)$')

    # plot posteriors for sigma
    ax3 = plt.subplot(222, sharey=ax1)
    ax3.plot(sig, psig, '-b')
    ax3.plot(sig, psig2, ':m')
    ax3.plot(sig, psig_norm, '--r')

    ax3.set_ylim(1E-4, 2)

    ax3.set_xlabel(r'$\sigma$')
    ax3.set_ylabel(r'$p(\sigma|x,I)$')

    ax4 = plt.subplot(224, sharex=ax3, sharey=ax2)
    ax4.plot(sig, psig.cumsum() * dsig, '-b')
    ax4.plot(sig, psig_norm.cumsum() * dsig, '--r')

    ax4.set_ylim(0, 1.05)
    ax4.set_xlim(0, 5)

    ax4.set_xlabel(r'$\sigma$')
    ax4.set_ylabel(r'$P(<\sigma|x,I)$')
```

```python
plot()
```

## 3. Gaussian Distribution with Gaussian Errors

Drawing samples from an unspecified Gaussian distribution $\mathcal{N}(\mu, \sigma)$.

In a more general case when the to be estimated quantity is drawn from such a distribution also has heteroscedastic errors.
$$
p\left(\left\{x_{i}\right\} | \mu, \sigma, I\right)=\prod_{i=1}^{N} \frac{1}{\sqrt{2 \pi}\left(\sigma^{2}+e_{i}^{2}\right)^{1 / 2}} \exp \left(\frac{-\left(x_{i}-\mu\right)^{2}}{2\left(\sigma^{2}+e_{i}^{2}\right)}\right)
$$

The log posterior probability becomes
$$
L_{p}=\text { constant }-\frac{1}{2} \sum_{i=1}^{N}\left(\ln \left(\sigma^{2}+e_{i}^{2}\right)+\frac{\left(x_{i}-\mu\right)^{2}}{\sigma^{2}+e_{i}^{2}}\right) $$

The underlying distribution is now not Gaussian but a weighted sum of Gaussians with varying widths, and thus the information content cannot be captured by three numbers.

Setting derivatives of $L_p$ w.r.t. $\mu$ and $\sigma$ equal to zero
$$ \mu_{0}=\frac{\sum_{i} w_{i} x_{i}}{\sum_{i} w_{i}}$$
$$\sum_{i=1}^{N} \frac{1}{\sigma_{0}^{2}+e_{i}^{2}}=\sum_{i=1}^{N} \frac{\left(x_{i}-\mu_{0}\right)^{2}}{\left(\sigma_{0}^{2}+e_{i}^{2}\right)^{2}}$$

where $$w_{i}=\frac{1}{\sigma_{0}^{2}+e_{i}^{2}}$$

Need to solve the coupled equations, better done numeriaclly.

```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
import numpy as np
from matplotlib import pyplot as plt
from astroML.plotting.mcmc import convert_to_stdev

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=False)


def gaussgauss_logL(xi, ei, mu, sigma):
    """Equation 5.63: gaussian likelihood with gaussian errors"""
    ndim = len(np.broadcast(sigma, mu).shape)

    xi = xi.reshape(xi.shape + tuple(ndim * [1]))
    ei = ei.reshape(ei.shape + tuple(ndim * [1]))

    s2_e2 = sigma ** 2 + ei ** 2
    return -0.5 * np.sum(np.log(s2_e2) + (xi - mu) ** 2 / s2_e2, 0)

#------------------------------------------------------------
# Define the grid and compute logL
np.random.seed(5)
mu_true = 1.
sigma_true = 1.
N = 10
ei = 3 * np.random.random(N)
xi = np.random.normal(mu_true, np.sqrt(sigma_true ** 2 + ei ** 2))

sigma = np.linspace(0.01, 5, 70)
mu = np.linspace(-3, 5, 70)

logL = gaussgauss_logL(xi, ei, mu, sigma[:, np.newaxis])
logL -= logL.max()

#------------------------------------------------------------
# plot the results
fig = plt.figure(figsize=(7, 5))
plt.imshow(logL, origin='lower',
           extent=(mu[0], mu[-1], sigma[0], sigma[-1]),
           cmap=plt.cm.binary,
           aspect='auto')
plt.colorbar().set_label(r'$\log(L)$')
plt.clim(-5, 0)

plt.text(0.5, 0.93,
         (r'$L(\mu,\sigma)\ \mathrm{for}\ \bar{x}=1,\ '
          r'\sigma_{\rm true}=1,\ n=10$'),
         bbox=dict(ec='k', fc='w', alpha=0.9),
         ha='center', va='center', transform=plt.gca().transAxes)

plt.contour(mu, sigma, convert_to_stdev(logL),
            levels=(0.683, 0.955, 0.997),
            colors='k')

plt.xlabel(r'$\mu$')
plt.ylabel(r'$\sigma$')

plt.show()
```
